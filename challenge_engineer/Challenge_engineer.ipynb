```python
import pandas as pd
```

```python
import requests

url = "https://api.stackexchange.com/2.2/search"
params = {
    "order": "desc",
    "sort": "activity",
    "intitle": "perl",
    "site": "stackoverflow"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data["items"])
    # Aquí puedes manipular los datos según tus necesidades
    print(data)
else:
    print("Error al hacer la solicitud:", response.status_code)
```

```python
df['creation_date'] = pd.to_datetime(df['creation_date'], unit='s')
df['last_activity_date'] = pd.to_datetime(df['last_activity_date'], unit='s')
df['last_edit_date'] = pd.to_datetime(df['last_edit_date'], unit='s')
```
